```python
import matplotlib.pyplot as plt
import numpy as np
import cv2
```

```python
image = cv2.imread('dataset/train/1++/cow_1++_1173.jpg')
image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

plt.imshow(image, cmap='gray')
```

```python
image[image > 10]  += 30
plt.imshow(image, cmap = 'gray')
```

```python
dst = cv2.equalizeHist(image)
plt.imshow(dst, cmap='gray')
```

```python
src = cv2.imread('dataset/train/1++/cow_1++_1173.jpg')

src_ycrcb = cv2.cvtColor(src, cv2.COLOR_BGR2YCrCb)
ycrcb_planes = list(cv2.split(src_ycrcb))


#      
ycrcb_planes[0] = cv2.equalizeHist(ycrcb_planes[0])

dst_ycrcb = cv2.merge(ycrcb_planes)
dst = cv2.cvtColor(dst_ycrcb, cv2.COLOR_YCrCb2RGB)

plt.imshow(dst)
```
